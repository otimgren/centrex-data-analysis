# Microwave leakage from RC to SPA 
Looking at the data we have that seems to indicate that microwaves from the rotational cooling region are leaking to the state preparation region, or even the detection region after that.

## Import packages

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import uncertainties as uc
```

## Set plot parameters

```python
plt.style.use(['seaborn-deep'])
# plt.style.use(['fivethirtyeight'])


# Font for labels
plt.rcParams['font.family'] = 'serif'

# Linewidths for plots
plt.rcParams['lines.linewidth'] = 1.6
plt.rcParams['lines.marker'] = 'x'

# Axes
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.grid'] = True


# Ticks
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16

# Legend
plt.rcParams['legend.fontsize'] = 16
```

## Save folder

```python
SAVE_DIR = "./saved_figs/RC_microwave_leakage/"
```

## Frequency scans
Start by getting data from file

```python
fname = ('G:\My Drive\CeNTREX Oskari\State preparation\SPA\Data analysis\Analyzed Data'
         '\SPA_test_11_1_2021_analyzed.hdf')

run_name_RC12 = "1635797009 RC J = 1 to 2 microwave power scan"
df_12 = pd.read_hdf(fname, run_name_RC12)

run_name_RC23 = "1635803356 Scan RC J = 2 to 3 microwave power"
df_23 = pd.read_hdf(fname, run_name_RC23)
```

## Plot signal size on R(1), F = 3

```python
fig, ax = plt.subplots(figsize = (16,9))
ax.errorbar(df_12.RC12_power, df_12.IntegratedFluorescence,
            yerr = df_12.IntegratedFluorescence_err)

ax.set_xlabel("RC J = 1 to 2 SynthHD power set point / dBm")
ax.set_ylabel("Fit integral (norm. by absorption)")
fig.savefig(SAVE_DIR+"sig_size_RC12_depl.pdf")
```

## Plot signal size on R(2), F = 4

```python
fig, ax = plt.subplots(figsize = (16,9))
ax.errorbar(df_23.RC23_power, df_23.IntegratedFluorescence,
            yerr = df_23.IntegratedFluorescence_err)

ax.set_xlabel("RC J = 2 to 3 SynthHD power set point / dBm")
ax.set_ylabel("Fit integral (norm. by absorption)")
fig.savefig(SAVE_DIR+"sig_size_RC23_depl.pdf")
```
